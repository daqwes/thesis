```python
import pandas as pd
```

```python
filename = "n_3_ns_2000_lambda_1000_eta_2.5e-05_matrix-type_rank-two_iter_5000_alpha_1.txt"
df = pd.read_csv(filename, header=None, names=["theta", "beta", "r", "err"])
valid_df = df[df["err"] != -1]
```

```python
valid_df.sort_values(by="err", ascending=True).head(20)#.groupby("theta").count()
```

For n=3 and rank-2, the combination 0.1, 100, 6 seems to be the best.

```python
filename4 = "n_4_ns_2000_lambda_1000_eta_2.5e-05_matrix-type_rank-two_iter_5000_alpha_1.txt"
df4 = pd.read_csv(filename4, header=None, names=["theta", "beta", "r", "err"])
valid_df4 = df4[(df4["err"] != -1) & (df4["err"] != 0)]
```

```python
valid_df4.sort_values(by="err", ascending=True).head(20)#.groupby("theta").count()
```

For n=4 and rank-2, the combination 1, 1000, 2 seems to be the best.
