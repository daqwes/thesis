```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = "iters.csv"

df_iters = pd.read_csv(file, index_col=0) 
```

```python
pl = df_iters[df_iters["type"] == "pl"]
prob = df_iters[df_iters["type"] == "prob"]
accs_pl = pl.acc
accs_prob = prob.acc
times_pl = pl.times
times_prob = prob.times
tup_prob = [(i, j) for i, j in zip(accs_prob, times_prob)]
tup_prob.sort(key=lambda x: x[1])
```

```python
plt.figure()
plt.plot(times_pl, accs_pl, label="langevin")
plt.plot(np.array(tup_prob)[:, 1], np.array(tup_prob)[:, 0], label="prob")
plt.legend()
plt.xlabel("Times [s]")
plt.ylabel("$L_2$ squared error")
plt.title("Comparison of accuracy wrt time")
plt.show()
```
